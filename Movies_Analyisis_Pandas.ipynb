```python
import pandas as pd
```

```python
df=pd.read_csv("movies.csv")
```

```python
df
```

```python
Total_Number_Of_Movies=df.title.size
print("Total Number Of Movies In The List:",Total_Number_Of_Movies)
```

```python
Total_Number_Of_Studios=df.studio.unique().size
print("Total Number Of Studio:",Total_Number_Of_Studios)
```

```python
Total_Number_Of_Languages=df.language.unique().size
print("Total Number Of Languages:",Total_Number_Of_Languages)
```

```python
Highest_IMDB_Rating=df[df.imdb_rating==df.imdb_rating.max()]
Highest_IMDB_Rating
```

```python
Lowest_IMDB_Rating=df[df.imdb_rating==df.imdb_rating.min()]
Lowest_IMDB_Rating
```

```python
Highest_Revenue_USD=df[df.currency=="USD"]
Highest_Revenue_USD[Highest_Revenue_USD.revenue==Highest_Revenue_USD.revenue.max()]
```

```python
Lowest_Revenue_Movie=df[df.unit=="Thousands"]
Lowest_Revenue_Movie[Lowest_Revenue_Movie.revenue==Lowest_Revenue_Movie.revenue.min()]
```

```python
Highest_Budget_Movie=df[df.currency=="USD"]
Highest_Budget_Movie[Highest_Budget_Movie.budget==Highest_Budget_Movie.budget.max()]
```

```python
Lowest_Budget_INR=df[df.currency=="INR"]
Lowest_Budget_INR[Lowest_Budget_INR.budget==Lowest_Budget_INR.budget.min()]
```

```python
df["Profit"]=df.apply(lambda x:x["revenue"]-x["budget"],axis=1)
df
```

```python
Highest_Profitable_Movie=df[df.currency=="USD"]
Highest_Profitable_Movie[Highest_Profitable_Movie.Profit==Highest_Profitable_Movie.Profit.max()]
```

```python
Lowest_Profitable_Movie=df[df.currency=="INR"]
Lowest_Profitable_Movie[Lowest_Profitable_Movie.Profit==Lowest_Profitable_Movie.Profit.min()]
```

```python
df["Profit_Percentage"]=df.apply(lambda x:(x["revenue"]-x["budget"])/x["budget"]*100,axis=1)
df
```

```python
Most_Profit_Percentage=df[df.Profit_Percentage==df.Profit_Percentage.max()]
Most_Profit_Percentage
```

```python
Least_Profit_Percentage=df[df.Profit_Percentage==df.Profit_Percentage.min()]
Least_Profit_Percentage
```

```python
idx=df[df.studio=="Marvel Studios"].imdb_rating.idxmax()
df.loc[idx]
```

```python
avg_HollyWood_IMDB=df[df.industry=="Hollywood"].imdb_rating.mean()
print("Average Hollywood Rating:",avg_HollyWood_IMDB)
```

```python
avg_BollyWood_IMDB=df[df.industry=="Bollywood"].imdb_rating.mean()
print("Average BollyWood Rating:",avg_BollyWood_IMDB)
```

```python
avg_HollyWood_Revenue=df[df.industry=="Hollywood"].revenue.mean()
print("Average Hollywood revenue:",avg_HollyWood_Revenue)
```

```python
avg_BollyWood_Revenue=df[df.industry=="Bollywood"].revenue.mean()
print("Average Bollywood revenue:",avg_BollyWood_Revenue)
```
