```python
df_complete_data=pd.concat([df_customers,df_order_items],axis=1)
df_complete_data
```

```python
df_complete_data=pd.merge(df_complete_data,df_reviews,on="order_id",how="left")
df_complete_data
```

```python
df_complete_data=pd.merge(df_complete_data,df_sellers,on="seller_id",how="left")
df_complete_data
```

```python
df_complete_data=pd.merge(df_complete_data,df_categories,on="product_category_name",how="left")
df_complete_data
```

```python
df_complete_data["total_amount_spent"] = df_complete_data.groupby("order_id")["payment_value"].transform("sum")
df_complete_data
```

```python
df_complete_data["order_purchase_timestamp"] = pd.to_datetime(df_complete_data["order_purchase_timestamp"])
df_complete_data["order_delivered_customer_date"] = pd.to_datetime(df_complete_data["order_delivered_customer_date"])
```

```python
df_complete_data["Delivery_Duration"]=df_complete_data.apply(lambda x:x["order_delivered_customer_date"]-x["order_purchase_timestamp"],axis=1)
df_complete_data
```

```python
def calcu(total):
    if total<=50:
        return "Low Spender"
    elif 50<total<=250:
        return "Medium Spender"
    else:
        return "High Spender"
df_complete_data["spending_category"]=df_complete_data.apply(lambda idx:calcu(idx['total_amount_spent']),axis=1)
df_complete_data
```

```python
df_complete_data["Delivery_Duration"].dtype
```

```python
def compare(time_val):
    if pd.isna(time_val):
        return 'unknown'
    if time_val<=pd.Timedelta(hours=3):
        return 'Fast'
    elif pd.Timedelta(hours=4)<=time_val<=pd.Timedelta(hours=7):
        return 'Normal'
    else:
        return 'Slow'
df_complete_data["Delivery_Category"]=df_complete_data.apply(lambda x:compare(x['Delivery_Duration']),axis=1)
df_complete_data                                                             
```

```python
def quality(rev):
    if rev==5:
        return 'Excellent'
    elif rev==4:
        return 'Good'
    elif rev==3:
        return 'Average'
    else:
        return 'Poor'
df_complete_data["review_quality"]=df_complete_data.apply(lambda x:quality(x['review_score']),axis=1)
df_complete_data
```

```python
import matplotlib.pyplot as plt

top_products = (
    df_complete_data.groupby("product_category_name_english")["order_item_id"]
    .count()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))
plt.bar(top_products.index, top_products.values)
plt.title("Top 10 Selling Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Number of Products Sold")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

payment = df_complete_data["payment_type"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    payment,
    labels=payment.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Payment Method Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

reviews = (
    df_complete_data["review_score"]
    .value_counts()
    .sort_index()
)

plt.figure(figsize=(8,5))
plt.bar(reviews.index.astype(str), reviews.values)
plt.title("Review Score Distribution")
plt.xlabel("Review Score")
plt.ylabel("Number of Reviews")
plt.tight_layout()
plt.show()
```
