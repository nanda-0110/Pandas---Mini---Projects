```python
import pandas as pd
```

### <font color="purple">**Question1**</font>
- Read the csv file with name "students_data.csv"
- Show the Top 5 rows

**Story:** ABC is maintaining the student records of a training institute. Read the student dataset and inspect the first five records.

```python
#read the csv file into the variable "df"
df=pd.read_csv("students_data.csv")

#show the top 5 rows
df.head(5)
```

### <font color="purple">**Question2**</font>
- Add a new column **performance**.
- If **score** is greater than or equal to 80, return **'Excellent'**, otherwise return **'Needs Improvement'**.

```python
#add the column "performance"
def performance(row):
    if row['score']>=80:
        return "Excellent"
    else:
        return "Needs Improvement"

df["performance"]=df.apply(performance,axis=1)

#show the df
df
```

### <font color="purple">**Question3**</font>

- Save the dataframe with the name **final_student_data.csv** and include only columns **student_id, name, course, score, performance**. Also set **index=False**.

```python
#save the dataframe
df.to_csv("final_student_data.csv",
          columns=["student_id","name","course","score","performance"],
          index=False)
```
